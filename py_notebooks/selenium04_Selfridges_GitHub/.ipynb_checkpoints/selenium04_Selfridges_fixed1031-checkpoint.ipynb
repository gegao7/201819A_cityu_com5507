```python
import unittest
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd 

driver = webdriver.Firefox() # please download the Firefox driver, and store it in the same folder of this code 
url = 'http://www.selfridges.com/US/en/cat/womens/bags/?browsing_country=US&currency=USD&language=en&pn=1&ppp=180'
driver.get(url)
#driver.maximize_window()
driver.implicitly_wait(2)

bagcontainer = pd.DataFrame(columns = ['name','desc','price']) 
result = driver.find_elements_by_xpath('//*[@id="searchTerm"]/div[2]/div/div') 
    # The "outer" table： //*[@id="searchTerm"]/div[2]/div                                
    # The first bag： //*[@id="searchTerm"]/div[2]/div/div[1]
    
    # xpath: //*[@id="searchTerm"]/div[2]/div/div[1]

for x in range(1,len(result)-3): # why -3? 
    name_element = driver.find_elements_by_xpath('//*[@id="searchTerm"]/div[2]/div/div['+str(x)+']/div/a[1]')[0]
    # //*[@id="searchTerm"]/div[2]/div/div[1]/div/a[1]
    name = name_element.text 
    
    desc_element = driver.find_elements_by_xpath('//*[@id="searchTerm"]/div[2]/div/div['+str(x)+']/div/a[2]')[0]
    desc = desc_element.text 

    price_element = driver.find_elements_by_xpath('//*[@id="searchTerm"]/div[2]/div/div['+str(x)+']/div/div[2]/ul/li/p')[0]
    price = price_element.text

    bagcontainer.loc[len(bagcontainer)] = [name,desc,price]

driver.close() 
```
